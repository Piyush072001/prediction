The inplace parameter in pandas functions determines whether the operation should modify the existing object (in-place) or return a new object with the changes applied. When inplace=True, the DataFrame is modified in place, and there is no need to assign the result back to a variable. If inplace=False or not specified (default), a new DataFrame with the changes is returned, and you need to assign it to a variable to capture the modified DataFrame.

```python
import re

# Assuming you have a DataFrame called 'df' with a column 'total_sqft'
# Replace 'total_sqft' with the actual column name in your DataFrame

# Function to convert various formats to single values
def convert_to_single(value):
    # Regular expression to match numeric values and units
    pattern = r'(\d+(\.\d+)?)\s*([a-zA-Z]+)?'
    
    match = re.match(pattern, str(value))
    if match:
        numeric_part, _, unit = match.groups()
        
        # Conversion factors for different units to Sq. Feet
        conversion_factors = {'Sq. Meter': 10.764, 'Perch': 272.25, 'Sq. Yards': 9}
        
        # Convert to Sq. Feet using the appropriate conversion factor
        return float(numeric_part) * conversion_factors.get(unit, 1)
    else:
        return float(value)

# Apply the function to the 'total_sqft' column
df_encoded['total_sqft'] = df_encoded['total_sqft'].apply(convert_to_single)
```

```python
df_encoded.head()
```

```python
def extract_numeric_from_size(value):
    # Regular expression to match numeric values
    pattern = r'(\d+(\.\d+)?)'
    
    match = re.search(pattern, str(value))
    if match:
        return float(match.group())
    else:
        return None  # Handle cases where no numeric part is found

# Apply the function to the 'size' column
df_encoded['size'] = df_encoded['size'].apply(extract_numeric_from_size)
```

```python
df_encoded.head()
df_encoded.dropna(inplace=True)
new_df=df_encoded
```

```python
x=new_df.drop('price',axis=1)
y=new_df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y ,test_size=0.2,random_state=10)
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
rmse = sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R2) Score: {r2}')
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(model, x, y, cv=5, scoring='r2')
```

```python
print(f'Mean R-squared score: {cv_scores.mean()}')
```
